```python
import pandas as pd
```

```python
df = pd.read_excel("DataV2/nba_odds_2018-19.xlsx")
df = df[["Date", "VH", "Team", "Open", "Close", "ML", "Final"]]
df
```

```python
games_df = pd.DataFrame(columns=["Date", "V", "H", "Open", "Close", "V_ML", "H_ML", "V_Score", "H_Score"])

for i in range(0, len(df), 2):
    g_df = df.loc[[i, i+1], :]
    
    row = {
        "Date": g_df.Date.iat[0],
        "V": g_df.Team.iat[0],
        "H": g_df.Team.iat[1],
        #"Open": g_df.Open.iat[line],
        #"Close": g_df.Close.iat[line],
        "V_ML": g_df.ML.iat[0],
        "H_ML": g_df.ML.iat[1],
        "V_Score": g_df.Final.iat[0],
        "H_Score": g_df.Final.iat[1],
    }
    
    for c in ["Open", "Close"]:
        if type(g_df[c].iat[0]) != str and g_df[c].iat[0] > 100:  # home favorite
            row[c] = -1 * g_df[c].iat[1] if type(g_df[c].iat[1]) != str else 0
        else:  # visitor favorite
            row[c] = g_df[c].iat[0] if type(g_df[c].iat[0]) != str else 0
    
    games_df = games_df.append(row, ignore_index=True)

games_df["Result"] = games_df["V_Score"] - games_df["H_Score"]
games_df["Open_Diff"] = games_df["Result"] - games_df["Open"]
games_df["Close_Diff"] = games_df["Result"] - games_df["Close"]

games_df
```

```python
games_df.to_csv("games_2018_2019.csv", index=False)
```
