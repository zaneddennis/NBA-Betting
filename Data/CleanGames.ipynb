```python
FALL = "2017"
SPRING = str(int(FALL)+1)
```

```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
df = pd.read_csv("lines_" + FALL + "_" + SPRING + ".tsv", sep="\t")
df.date = df.date.astype("str")
df
```

```python
def reformatDate(date):
    dow, m, d, y = date.split(" ")
    
    if int(d) < 10:
        d = "0" + d
    mDict = {"Oct": "10", "Nov": "11", "Dec": "12", "Jan": "01", "Feb": "02", "Mar": "03", "Apr": "04"}
    
    return y + mDict[m] + d

def translateTeam(team):
    splits = team.split(" ")
    
    if team == "Portland Trail Blazers":
        return "Portland"
    elif team == "Los Angeles Lakers":
        return "L.A. Lakers"
    elif team == "Los Angeles Clippers":
        return "L.A. Clippers"
    else:
        return " ".join(splits[:-1])

scores_df = pd.DataFrame()
for month in ["oct", "nov", "dec"]:
    month_df = pd.read_csv("scores_" + month + "_" + FALL + ".csv")
    
    month_df.date = month_df.date.apply(reformatDate)
    month_df.away = month_df.away.apply(translateTeam)
    month_df.home = month_df.home.apply(translateTeam)

    scores_df = scores_df.append(month_df)

for month in ["jan", "feb", "mar", "apr"]:
    month_df = pd.read_csv("scores_" + month + "_" + SPRING + ".csv")
    
    month_df.date = month_df.date.apply(reformatDate)
    month_df.away = month_df.away.apply(translateTeam)
    month_df.home = month_df.home.apply(translateTeam)

    scores_df = scores_df.append(month_df)

df = pd.merge(df, scores_df, how="left", on=["date", "away", "home"])
df
#df[df.isnull().any(axis=1)]
```

```python
df["result"] = df.away_pts - df.home_pts
df["result_diff"] = df["result"] - df["opening_spread"]
df
```

```python
def convertDate(date):
    day0 = min(df.date)
    day0 = datetime.strptime(day0, "%Y%m%d")
    dayCurr = datetime.strptime(date, "%Y%m%d")
    delta = dayCurr - day0
    
    return delta.days

df["day"] = df.date.apply(convertDate)
df = df[["date", "day", "away", "home", "opening_spread", "away_ML", "home_ML", "away_pts", "home_pts", "result", "result_diff"]]
df
```

```python
df.to_csv("games_cleaned_" + FALL + "_" + SPRING + ".tsv", sep="\t", index=False)
```
