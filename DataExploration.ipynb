```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
HCA = 3.0
```

```python
df = pd.read_csv("DataV2/games_2018_2019.csv")
df = df.loc[:1229]  # remove playoffs
df
```

```python
srs_df = pd.read_csv("DataV2/srs_2019.csv", header=None)
srs_df = srs_df.set_index(0)
srs_df.columns=["SRS"]
srs_df
```

```python
df["V_SRS"] = 0.0
df["H_SRS"] = 0.0
for index, row in df.iterrows():
    df.at[index, "V_SRS"] = srs_df.at[row["V"], "SRS"]
    df.at[index, "H_SRS"] = srs_df.at[row["H"], "SRS"]

df["V_GameScore"] = df["H_SRS"] + df["Result"] + HCA  # TODO: come up with a better HCA system
df["H_GameScore"] = df["V_SRS"] - df["Result"] - HCA

df
```

```python
teams = list(df.sort_values(by="V").V.unique())
teams
```

```python
for t in teams:
    print("{}: {}".format(t, srs_df.at[t, "SRS"]))
    v_df = df.loc[df.V==t]
    h_df = df.loc[df.H==t]
    assert len(v_df) == len(h_df)
    
    print("V: {}".format(v_df.V_GameScore.mean() - HCA))
    print("H: {}".format(h_df.H_GameScore.mean() + HCA))
    print("=========")
    all_GS = v_df.V_GameScore.append(h_df.H_GameScore)
    print("Tot: {}".format(all_GS.mean()))
    print("SD: {}".format(all_GS.std()))
    print("Sk: {}".format(all_GS.skew()))
    print("Ku: {}".format(all_GS.kurtosis()))
    
    sns.distplot(v_df.V_GameScore.append(h_df.H_GameScore))
    plt.show()
    
    print()
```
